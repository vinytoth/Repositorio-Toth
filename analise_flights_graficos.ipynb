# Análise Exploratória com o Dataset `flights` (Seaborn)

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

df = sns.load_dataset("flights")
df.head()
```

```python

sns.lineplot(data=df, x="year", y="passengers", ci=None, marker="o")
plt.title("Tendência de Passageiros ao Longo dos Anos")
plt.ylabel("Nº de Passageiros")
plt.show()
```

```python

sns.boxplot(data=df, x="month", y="passengers")
plt.title("Distribuição Mensal de Passageiros")
plt.xticks(rotation=45)
plt.show()
```

```python

pivot = df.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Heatmap de Passageiros por Mês e Ano")
plt.show()
```
